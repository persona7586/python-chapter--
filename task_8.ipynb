```python
#mymath.py 모듈을 이용한 내용은 이렇게 작성 해야되지만 에러가 발생해서 변수 지정식으로 아래와 같이 풀어봤음;

#mymath.py 파일 안에 있어야 할 내용
import math

def triangle_area(base, height):
    return 0.5 * base * height

def circle_area(radius):
    return math.pi * radius ** 2

def prism_area(length, width, height):
    return 2 * (length * width + length * height + width * height)

#main.py 파일 안에 있어야 할 내용
import mymath

def main():
    choice = input('다음 중 구하고 싶은 값을 선택하십시5 \n 1.삼각형 넓이 \n 2.원의 넓이 \n 3.직육면체의 넓이\n')

    if choice == '1':
        triangle_area_input = input('삼각형의 넓이를 구해드리겠소 밑변과 높이를 입력해 주십시5 (예: 5 7): ').split()
        if len(triangle_area_input) != 2:
            print('제대로 입력 해주십시5 ex) 5 7')
        else:
            base, height = map(float, triangle_area_input)
            if base > 0 and height > 0:
                area = mymath.triangle_area(base, height)
                print(f'삼각형의 넓이 : {area}')
            else:
                print('음수와 0은 입력하지 마십시5')

    elif choice == '2':
        circle_area_input = input('원의 넓이를 구해드리겠소 반지름을 입력해 주십시5 (예: 5): ').split()
        if len(circle_area_input) != 1:
            print('제대로 입력 해주십시5 ex) 5')
        else:
            radius = float(circle_area_input[0])
            if radius > 0:
                area = mymath.circle_area(radius)
                print(f'원의 넓이 : {area:.2f}')
            else:
                print('음수와 0은 입력하지 마십시5')

    elif choice == '3':
        prism_area_input = input('직육면체의 표면적을 구해드리겠소 길이, 너비, 높이를 순서대로 입력해 주십시5 (예: 7 17 27): ').split()
        if len(prism_area_input) != 3:
            print('제대로 입력 해주십시5 ex) 7 17 27')
        else:
            length, width, height = map(float, prism_area_input)
            if length > 0 and width > 0 and height > 0:
                area = mymath.prism_area(length, width, height)
                print(f'직육면체의 표면적 : {area}')
            else:
                print('음수와 0은 입력하지 마십시5')

    else:
        print('잘못된 선택입니다.')

if __name__ == '__main__':
    main()
```

```python
#미리 변수를 지정하고 그 안에 호출 수식 내용을 남겨놓음
def triangle_area(base, height): #필요한 수식: 밑변, 높이
    return 0.5 * base * height #밑변과 높이를 이용해 삼각형의 넓이 구하는 공식

def circle_area(radius): #필요한 수식: 반지름
    import math      #math 모듈을 가져와 사용할 때 필요한 구문으로 math는 수학 공식과 관련된 다양한 기능을 제공해줌
    return math.pi * radius ** 2 #반지름을 이용해 원의 넓이를 구하는 공식

def prism_area(length, width, height): #필요한 수식: 길이, 너비, 높이
    return 2 * (length * width + length * height + width * height) #길이, 너비, 높이를 이용해 직육면체의 넓이를 구하는 공식

def main(): #main() 함수는 프로그램의 주요 작업을 수행하는 함수로 프로그램의 시작점이자 실행되는 함수로 사용

    # 사용자가 선택해서 정할 수 있게 하는 메뉴 출력
    choice = input('다음 중 구하고 싶은 값을 선택하십시5 \n 1.삼각형 넓이 \n 2.원의 넓이 \n 3.직육면체의 넓이\n') #출력값이 세로로 나올 수 있도록 역슬레쉬n을 이용해서 내림

    if choice == '1': #첫번째 선택은 삼각형의 넓이를 구하는 내용을 입력했고 두개의 값이 입력되어야 하며 다른걸 입력했을때와 음수와 0을 입력했을 때 정정 안내를 첨부함
        triangle_area_input = input('삼각형의 넓이를 구해드리겠소 밑변과 높이를 입력해 주십시5 (예: 5 7): ').split()
        if len(triangle_area_input) != 2:
            print('제대로 입력 해주십시5 ex) 5 7')
        else:
            base, height = map(float, triangle_area_input)
            if base > 0 and height > 0:
                area = triangle_area(base, height)
                print(f'삼각형의 넓이 : {area}')
            else:
                print('음수와 0은 입력하지 마십시5')

    elif choice == '2': #두번째 선택은 원의 넓이를 구하는 내용을 입력했고 한개의 값이 입력되어야 하며 다른걸 입력했을때와 음수와 0을 입력했을 때 정정 안내를 첨부함
        circle_area_input = input('원의 넓이를 구해드리겠소 반지름을 입력해 주십시5 (예: 5): ').split()
        if len(circle_area_input) != 1:
            print('제대로 입력 해주십시5 ex) 5')
        else:
            radius = float(circle_area_input[0])
            if radius > 0:
                area = circle_area(radius)
                print(f'원의 넓이 : {area:.2f}') #원주율이 계산되는 순간부터 소숫점이 길어질 것이므로 소숫점이 길어지지 않도록 .2f = 소숫점 둘째 자리까지 표현 설정해줌
            else:
                print('음수와 0은 입력하지 마십시5')

    elif choice == '3': #세번째 선택은 직육면체의 표면적을 구하는 내용을 입력했고 세개의 값이 입력되어야 하며 다른걸 입력했을때와 음수와 0을 입력했을 때 정정 안내를 첨부함
        prism_area_input = input('직육면체의 표면적을 구해드리겠소 길이, 너비, 높이를 순서대로 입력해 주십시5 (예: 7 17 27): ').split()
        if len(prism_area_input) != 3:
            print('제대로 입력 해주십시5 ex) 7 17 27')
        else:
            length, width, height = map(float, prism_area_input)
            if length > 0 and width > 0 and height > 0:
                area = prism_area(length, width, height)
                print(f'직육면체의 표면적 : {area}')
            else:
                print('음수와 0은 입력하지 마십시5')

    else:
        print('잘못된 선택입니다.')

if __name__ == '__main__': #if __name__ == '__main__':를 사용하면, 프로그램이 메인 스크립트로 실행될 때만 main() 함수가 실행됨. 다른 파일에서 이 스크립트를 임포트할 때는 main() 함수가 실행되지 않으므로, 코드 재사용성을 높일 수 있음.
    main() #함수 안에 프로그램의 주요 로직을 넣어두면, 프로그램의 흐름을 더 명확하게 파악할 수 있음.
```
